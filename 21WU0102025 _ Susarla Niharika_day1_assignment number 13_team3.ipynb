Question 1: Create two int type variables, apply addition, subtraction, division and multiplications and store the results in variables. Then print the data in the following format by calling the variables:

 First variable is __ & second variable is __.

 Addition: __ + __ = __

 Subtraction: __- __ = __

 Multiplication: __ * __ = __

 Division: __ / __ = __

```python

a = 10
b = 5

addition = a + b
subtraction = a - b
multiplication = a * b
division = a / b

print(f"First variable is {a} & second variable is {b}.")
print(f"Addition: {a} + {b} = {addition}")
print(f"Subtraction: {a} - {b} = {subtraction}")
print(f"Multiplication: {a} * {b} = {multiplication}")
print(f"Division: {a} / {b} = {division}")
```

Question 2: What is the difference between the following operators:

 (i) '/' & '//'

 (ii) ‘**’ & ‘^’

 Answer:  **(i) '/' vs. '//':**  

- **`/`**: Performs **floating-point division**, always returns a float.

  Example: `5 / 2 = 2.5`

- **`//`**: Performs **floor division**, returns the largest integer less than or equal to the result.

  Example: `5 // 2 = 2`

**(ii) '**' vs. '^':**  

- **`**`**: Performs **exponentiation**  

  Example: `2 ** 3 = 8`

- **`^`**: Performs **bitwise XOR**  between two numbers.  

  Example: `2 ^ 3 = 1`

**Question 3:  List the logical operators**

Answer: The logical operators in Python are:

1. **`and`**: Returns `True` if both conditions are true.

2. **`or`**: Returns `True` if at least one condition is true.

3. **`not`**: Reverses the boolean value (True to False, False to True).

**Question 4:  Explain right shift operator and left shift operator with examples**

Answer: **Right Shift (>>)**: Moves the bits of a number to the right dividing the number by powers of 2.

Example:

x = 8  

result = x >> 2

print(result)

**Left Shift (<<)**: Moves the bits of a number to the left, effectively multiplying the number by powers of 2.

Example:

x = 3

result = x << 2

print(result)
