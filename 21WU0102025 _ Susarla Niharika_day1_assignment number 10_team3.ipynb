**Question 1:  What is the role of try and exception block?**

Answer: The try block is used to write code that might cause an error. The except block handles the error if it occurs, preventing the program from crashing

**Question 2:  What is the syntax for a basic try-except block?**

```python
try:
    # Code that might raise an exception
except ExceptionType:
    # Code to handle the exception
```

**Question 3: What happens if an exception occurs inside a try block and there is no matching except block?**

Answer: If an exception occurs inside a try block and there is no matching except block, the program stops execution the exception propagates up the call stack and Python prints an error message with a traceback. If no handler is found, the program terminates

**Question 4: What is the difference between using a bare except block and specifying a specific exception type?**

Answer: Using a bare except block:

- Catches all exceptions including system-related ones like KeyboardInterrupt or SystemExit  
- Makes it harder to debug and may hide unexpected errors

Using a specific exception type:

- Catches only the specified exception, like ValueError  
- Ensures only expected errors are handled, making the code safer and more predictable

**Question 5:  Can you have nested try-except blocks in Python? If yes, then give an example**

Answer: Yes python allows nested try-except blocks

```python
Example: try:
    num = int(input("Enter a number: "))
    try:
        print(10 / num)
    except ZeroDivisionError:
        print("Cannot divide by zero!")
except ValueError:
    print("Invalid input! Please enter a number.")
```

**Question 6:Can we use multiple exception blocks, if yes then give an example.**

Answer: Yes you can use multiple except blocks to handle different exceptions


```python
Example: try:
    num = int(input("Enter a number: "))
    print(10 / num)
except ValueError:
    print("Invalid input! Please enter a number.")
except ZeroDivisionError:
    print("Cannot divide by zero!")
```

Question 7:  Write the reason due to which following errors are raised:

 a. EOFError

 b. FloatingPointError

 c. IndexError

 d. MemoryError

 e. OverflowError

 f.TabError

 g. ValueError

 Answer:

**a. `EOFError`:** Raised when `input()` hits the end of the file without reading any data

**b. `FloatingPointError`:** Raised when a floating-point operation fails (rare, often platform-specific)

**c. `IndexError`:** Raised when trying to access an invalid index in a sequence (e.g., list, tuple)  

**d. `MemoryError`:** Raised when an operation runs out of memory

**e. `OverflowError`:** Raised when a calculation exceeds the maximum limit for a numeric type

**f. `TabError`:** Raised when inconsistent tabs and spaces are used in indentation  

**g. `ValueError`:** Raised when a function receives an argument of the correct type but with an invalid value

Question 8: Write code for the following given scenario and add try-exception block to it.

 a. Program to divide two numbers

 b. Program to convert a string to an integer

 c. Program to access an element in a list

 d. Program to handle a specific exception

 e. Program to handle any exception


```python
a. Divide two numbers

try:
    print(float(input("Enter numerator: ")) / float(input("Enter denominator: ")))
except ZeroDivisionError:
    print("Cannot divide by zero!")
except ValueError:
    print("Invalid input!")

```

```python
b. Convert a string to an integer

try:
    print(int(input("Enter a number: ")))
except ValueError:
    print("Invalid input!")
```

```python
c. Access an element in a list

try:
    items = [1, 2, 3]
    print(items[int(input("Enter index: "))])
except IndexError:
    print("Index out of range!")
except ValueError:
    print("Invalid input!")
```

```python
d. Handle a specific exception

try:
    print(10 / int(input("Enter a number: ")))
except ZeroDivisionError:
    print("Cannot divide by zero!")
```

```python
e. Handle any exception

try:
    print(10 / int(input("Enter a number: ")))
except Exception as e:
    print(f"Error: {e}")
```
