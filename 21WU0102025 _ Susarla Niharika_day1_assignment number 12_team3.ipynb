**Question 2:  What are the rules to create variables in python?**

Answer:  Rules to create variables in Python are:

1. It will start with a letter or underscore

2. It can contain letters, numbers, and underscores

3. It is case-sensitive (e.g., `myVar` and `myvar` are different)

4.  It cannot use Python keywords

5. Avoid starting with a number

**Question 3: What are the standards and conventions followed for the nomenclature of variables in  python to improve code readability and maintainability?**

Answer: In python variable names should be in lowercase with words separated by underscores (e.g., `my_variable`). Using descriptive names to clarify their purpose (e.g., `user_age`). Constants are written in uppercase (e.g., `MAX_SIZE`). Avoid single-character names unless in loops (e.g., `i`, `j`). Never use python keywords as variable names (e.g., `if`, `class`).

**Question 4: What will happen if a keyword is used as a variable name?**

**Question 5: For what purpose def keyword is used?**

**Question 6: What is the operation of this special character ‘\’?**

Answer: The special character **`\`** is called the **escape character** in Python. It is used to introduce special character sequences, like:

- **`\n`**: Newline

- **`\t`**: Tab

- **`\\`**: Backslash

It allows you to include characters that are otherwise difficult to type or use in strings
