**Question 1: What is the role of the 'else' block in a try-except statement? Provide an example scenario where it would be useful.**

Answer: The else block in a try-except statement runs only if no exception occurs in the try block. If no exception occurs the result is printed in the else block

```python
try:
    num1 = int(input("Enter numerator: "))
    num2 = int(input("Enter denominator: "))
    result = num1 / num2
except ZeroDivisionError:
    print("Cannot divide by zero!")
except ValueError:
    print("Invalid input!")
else:
    print(f"Result: {result}")
```

**Question 2:  Can a try-except block be nested inside another try-except block? Explain with an example. short and simple**

Answer: Yes a try-except block can be nested inside another try-except block. This allows you to handle different exceptions at different levels

```python
Example:

try:
    num1 = int(input("Enter numerator: "))
    try:
        num2 = int(input("Enter denominator: "))
        result = num1 / num2
    except ZeroDivisionError:
        print("Cannot divide by zero!")
except ValueError:
    print("Invalid input!")
```

**Question 3: How can you create a custom exception class in Python? Provide an example that demonstrates its usage.**

Answer: You can create a custom exception class in Python by inheriting from the built-in Exception class

MyCustomError is a custom exception class

It is raised in the try block and caught in the except block

```python
Example:

class MyCustomError(Exception):
    pass

try:
    raise MyCustomError("This is a custom exception")
except MyCustomError as e:
    print(f"Caught an error: {e}")
```

Answer: Logging in Python tracks events during program execution. It will help with debugging, monitoring, and error tracking. The logging module provides flexible logging levels (example: INFO, ERROR). It is important for identifying issues monitoring performance and maintaining logs for auditing

**Question 6:  Explain the purpose of log levels in Python logging and provide examples of when each log level would be appropriate.**

Answer: Log levels in Python help control the severity of messages in logs

1. **`DEBUG`**: Detailed information for debugging.  
   Example: `logging.debug("User input received")`

2. **`INFO`**: General program information.  
   Example: `logging.info("Process completed successfully")`

3. **`WARNING`**: Potential issues to watch.  
   Example: `logging.warning("Memory usage high")`

4. **`ERROR`**: Errors affecting functionality.  
   Example: `logging.error("Failed to connect to database")`

5. **`CRITICAL`**: Severe issues that may stop the program.  
   Example: `logging.critical("System crash")`

**Question 7:  What are log formatters in Python logging, and how can you customise the log message format using formatters?**

Answer: Log formatters in Python define how log messages are structured. We can customize the log format using the format parameter in logging.basicConfig() or with a Formatter object. Common format codes include %(asctime)s for timestamp, %(levelname)s for log level, and %(message)s for the log message. Customizing the format helps organize and make logs more readable

```python
Example:

logging.basicConfig(format='%(asctime)s - %(levelname)s - %(message)s', level=logging.DEBUG)
logging.info("Info message")
```

**Question 8:  How can you set up logging to capture log messages from multiple modules or classes in a Python application?**

Answer:
To capture log messages from multiple modules, set up logging in the main script and use logging.getLogger(__name__) in each module. This ensures all modules use the same logging configuration.

```python
Example:

Main script:

import logging
logging.basicConfig(level=logging.DEBUG)

Module:

import logging
logger = logging.getLogger(__name__)
logger.info("Log from module")
```

Question 10:  10. Write a Python program that logs a message to a file named "app.log" with the following requirements:

 ● Thelogmessage should be "Hello, World!"

 ● Theloglevel should be set to "INFO."

 ● Thelogfile should append new log entries without overwriting previous ones.

```python
import logging

# Set up logging to append to the file
logging.basicConfig(filename='app.log', level=logging.INFO, format='%(asctime)s - %(message)s', filemode='a')

# Log the message
logging.info("Hello, World!")

This will append the log message with the INFO level to app.log without overwriting existing entries
```

**Question 11:  Create a Python program that logs an error message to the console and a file named "errors.log" if an exception occurs during the program's execution. The error message should include the exception type and a timestamp**
